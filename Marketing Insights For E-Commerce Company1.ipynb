```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.feature_selection import f_classif # For F-score (ANOVA)
from statsmodels.stats.outliers_influence import variance_inflation_factor # For VIF
from sklearn.decomposition import PCA # For PCA

# Set matplotlib style for better aesthetics
plt.style.use('seaborn-v0_8-darkgrid')

# Load the datasets
try:
    cust_data = pd.read_excel(r"C:\Users\DELL\OneDrive\Desktop\rahul\14. Capstone Case Study - Finding-Marketing-Insights\CustomersData.xlsx")
    discount_coupon = pd.read_excel(r"C:\Users\DELL\OneDrive\Desktop\rahul\14. Capstone Case Study - Finding-Marketing-Insights\Discount_Coupon.xlsx")
    marketing_spend = pd.read_excel(r"C:\Users\DELL\OneDrive\Desktop\rahul\14. Capstone Case Study - Finding-Marketing-Insights\Marketing_Spend.xlsx")
    online_sales = pd.read_excel(r"C:\Users\DELL\OneDrive\Desktop\rahul\14. Capstone Case Study - Finding-Marketing-Insights\Online_Sales.xlsx")
    tax_amt = pd.read_excel(r"C:\Users\DELL\OneDrive\Desktop\rahul\14. Capstone Case Study - Finding-Marketing-Insights\Tax_amount.xlsx")
except FileNotFoundError as e:
    print(f"Error loading file: {e}. Please ensure all CSV files are in the same directory.")
    # Exit or handle the error appropriately
    exit()

# --- Data Preprocessing ---

# Clean column names by removing leading/trailing spaces
discount_coupon.columns = discount_coupon.columns.str.strip()
marketing_spend.columns = marketing_spend.columns.str.strip()
online_sales.columns = online_sales.columns.str.strip()
tax_amt.columns = tax_amt.columns.str.strip()
cust_data.columns = cust_data.columns.str.strip()

# Convert 'Transaction_Date' in online_sales to datetime objects
online_sales['Transaction_Date'] = pd.to_datetime(online_sales['Transaction_Date'], format='%Y%m%d')
# Extract month abbreviation and Year-Month for analysis
online_sales['Month'] = online_sales['Transaction_Date'].dt.strftime('%b')
online_sales['Year_month'] = online_sales['Transaction_Date'].dt.strftime('%Y-%m')
online_sales['Day_of_Week'] = online_sales['Transaction_Date'].dt.day_name()
online_sales['Day'] = online_sales['Transaction_Date'].dt.day

# Convert 'Date' in marketing_spend to datetime objects
marketing_spend['Date'] = pd.to_datetime(marketing_spend['Date'], format='%m/%d/%Y')
marketing_spend['Year_month'] = marketing_spend['Date'].dt.strftime('%Y-%m')
marketing_spend['Total_spend'] = marketing_spend['Offline_Spend'] + marketing_spend['Online_Spend']

# --- Merge DataFrames ---
# Merge online_sales with tax_amt on 'Product_Category'
online_sales_merged = pd.merge(online_sales, tax_amt, how='left', on='Product_Category')

# Merge the result with cust_data on 'CustomerID'
online_sales_merged = pd.merge(online_sales_merged, cust_data, how='left', on='CustomerID')

# --- CORRECTED MERGE FOR DISCOUNT_COUPON ---
# Create a common key for merging discount_coupon
online_sales_merged['Month_Product_Key'] = online_sales_merged['Month'] + '_' + online_sales_merged['Product_Category']
discount_coupon['Month_Product_Key'] = discount_coupon['Month'] + '_' + discount_coupon['Product_Category']

# Perform a left merge to bring in Discount_pct from discount_coupon.
# The column from 'discount_coupon' will be directly named 'Discount_pct'.
online_sales_merged = pd.merge(online_sales_merged, discount_coupon[['Month_Product_Key', 'Discount_pct']],
                               how='left', on='Month_Product_Key')

# Now, handle the final Discount_pct based on Coupon_Status and fill NaNs.
# If Coupon_Status is 'Not Used', the discount should be 0.
online_sales_merged.loc[online_sales_merged['Coupon_Status'] == 'Not Used', 'Discount_pct'] = 0

# Fill any remaining NaNs in Discount_pct with 0.
online_sales_merged['Discount_pct'].fillna(0, inplace=True)

# Drop the temporary merge key
online_sales_merged.drop(columns=['Month_Product_Key'], errors='ignore', inplace=True)


final_data = online_sales_merged.copy()

# --- 1. Calculate Invoice amount or sale_amount or revenue for each transaction and item level ---
# Invoice Value =(( Quantity*Avg_price)*(1-Dicount_pct)*(1+GST))+Delivery_Charges
final_data['Invoice_amt'] = (
    (final_data['Quantity'] * final_data['Avg_Price']) *
    (1 - final_data['Discount_pct'] / 100) *
    (1 + final_data['GST'])
) + final_data['Delivery_Charges']

# Sale_amount (Revenue before GST and Delivery Charges, after discount)
final_data['Sales_amt'] = (final_data['Quantity'] * final_data['Avg_Price']) * (1 - final_data['Discount_pct'] / 100)

print("--- Sample of final_data with calculated Invoice_amt and Sales_amt ---")
print(final_data[['Transaction_ID', 'Product_Description', 'Quantity', 'Avg_Price', 'Discount_pct', 'GST', 'Delivery_Charges', 'Invoice_amt', 'Sales_amt']].head())
print("\n" + "="*80 + "\n")

# --- 2. Perform Detailed Exploratory Analysis ---

# --- Understanding how many customers acquired every month ---
# Get the first transaction date for each customer
first_purchase = final_data.groupby('CustomerID')['Transaction_Date'].min().reset_index()
first_purchase['Acquisition_Month'] = first_purchase['Transaction_Date'].dt.strftime('%Y-%m')

monthly_acquisition = first_purchase.groupby('Acquisition_Month').size().reset_index(name='New_Customers')

print("--- New Customers Acquired Every Month ---")
print(monthly_acquisition)
plt.figure(figsize=(12, 6))
sns.barplot(x='Acquisition_Month', y='New_Customers', data=monthly_acquisition, palette='viridis')
plt.title('New Customers Acquired Every Month')
plt.xlabel('Month')
plt.ylabel('Number of New Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# --- Understand the retention of customers on month on month basis ---
# Create a 'CohortGroup' which is the month of the first purchase
final_data['CohortGroup'] = final_data.groupby('CustomerID')['Transaction_Date'].transform('min').dt.strftime('%Y-%m')

# Create 'CohortPeriod' which is the number of months since the first purchase
def get_cohort_period(df):
    df['CohortPeriod'] = (df['Transaction_Date'].dt.to_period('M') - df['CohortGroup'].astype('period[M]')).apply(lambda x: x.n)
    return df

# Apply the function and then reset the index to ensure CustomerID is a column, not an index level
final_data = final_data.groupby('CustomerID', group_keys=False).apply(get_cohort_period).reset_index()

# Calculate total users in each cohort and their retention
cohorts = final_data.groupby(['CohortGroup', 'CohortPeriod'])['CustomerID'].nunique().reset_index()
cohorts.rename(columns={'CustomerID': 'Total_Users'}, inplace=True)

cohort_group_size = cohorts[cohorts['CohortPeriod'] == 0].set_index('CohortGroup')['Total_Users']
cust_retention = cohorts.pivot_table(index='CohortGroup', columns='CohortPeriod', values='Total_Users')
cust_retention = cust_retention.divide(cohort_group_size, axis=0) * 100 # Convert to percentage

print("--- Customer Retention by Monthly Cohorts (%) ---")
print(cust_retention.round(2))
plt.figure(figsize=(15, 8))
sns.heatmap(cust_retention, annot=True, fmt='.1f', cmap='Greens', vmin=0, vmax=cust_retention.max().max() + 5)
plt.title('Retention by Monthly Cohorts (%)')
plt.ylabel('Cohort Group (Acquisition Month)')
plt.xlabel('Cohort Period (Months Since Acquisition)')
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# --- How the revenues from existing/new customers on month on month basis ---
# Identify new vs. existing customers for each transaction
final_data['Is_New_Customer'] = final_data.apply(
    lambda row: 1 if row['Transaction_Date'].strftime('%Y-%m') == row['CohortGroup'] else 0, axis=1
)

monthly_revenue_type = final_data.groupby(['Year_month', 'Is_New_Customer'])['Sales_amt'].sum().unstack(fill_value=0)
monthly_revenue_type.rename(columns={0: 'Existing_Customer_Revenue', 1: 'New_Customer_Revenue'}, inplace=True)

print("--- Monthly Revenue from New vs. Existing Customers ---")
print(monthly_revenue_type)
monthly_revenue_type.plot(kind='bar', stacked=True, figsize=(14, 7), cmap='coolwarm')
plt.title('Monthly Revenue from New vs. Existing Customers')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# --- How the discounts playing role in the revenues? ---
# Highlight Start: Changes made in this block
discount_revenue = final_data.groupby('Discount_pct')['Sales_amt'].sum().reset_index()

print("--- Sales Amount by Discount Percentage ---")
print(discount_revenue)
plt.figure(figsize=(10, 6))
# Fix for FutureWarning: Passing `palette` without assigning `hue` is deprecated
sns.barplot(x='Discount_pct', y='Sales_amt', data=discount_revenue, palette='rocket', hue='Discount_pct', legend=False)
plt.title('Total Sales Amount by Discount Percentage')
plt.xlabel('Discount Percentage (%)')
plt.ylabel('Total Sales Amount')
plt.tight_layout()
plt.show()
# Highlight End: Changes made in this block
print("\n" + "="*80 + "\n")

# --- Analyse KPIs like Revenue, number of orders, average order value, number of customers (existing/new), quantity, by category, by month, by week, by day etc ---

# KPIs by Product Category
kpis_by_category = final_data.groupby('Product_Category').agg(
    Total_Revenue=('Sales_amt', 'sum'),
    Number_of_Orders=('Transaction_ID', 'nunique'),
    Total_Quantity=('Quantity', 'sum'),
    Number_of_Customers=('CustomerID', 'nunique')
).reset_index()
kpis_by_category['Average_Order_Value'] = kpis_by_category['Total_Revenue'] / kpis_by_category['Number_of_Orders']

print("--- KPIs by Product Category ---")
print(kpis_by_category.round(2))
print("\n" + "="*80 + "\n")

# KPIs by Month
kpis_by_month = final_data.groupby('Year_month').agg(
    Total_Revenue=('Sales_amt', 'sum'),
    Number_of_Orders=('Transaction_ID', 'nunique'),
    Total_Quantity=('Quantity', 'sum'),
    Number_of_Customers=('CustomerID', 'nunique')
).reset_index()
kpis_by_month['Average_Order_Value'] = kpis_by_month['Total_Revenue'] / kpis_by_month['Number_of_Orders']

print("--- KPIs by Month ---")
print(kpis_by_month.round(2))
print("\n" + "="*80 + "\n")

# KPIs by Day of Week
kpis_by_day_of_week = final_data.groupby('Day_of_Week').agg(
    Total_Revenue=('Sales_amt', 'sum'),
    Number_of_Orders=('Transaction_ID', 'nunique'),
    Total_Quantity=('Quantity', 'sum'),
    Number_of_Customers=('CustomerID', 'nunique')
).reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).reset_index()
kpis_by_day_of_week['Average_Order_Value'] = kpis_by_day_of_week['Total_Revenue'] / kpis_by_day_of_week['Number_of_Orders']

print("--- KPIs by Day of Week ---")
print(kpis_by_day_of_week.round(2))
print("\n" + "="*80 + "\n")

# KPIs by Day (of month)
kpis_by_day = final_data.groupby('Day').agg(
    Total_Revenue=('Sales_amt', 'sum'),
    Number_of_Orders=('Transaction_ID', 'nunique'),
    Total_Quantity=('Quantity', 'sum'),
    Number_of_Customers=('CustomerID', 'nunique')
).reset_index()
kpis_by_day['Average_Order_Value'] = kpis_by_day['Total_Revenue'] / kpis_by_day['Number_of_Orders']

print("--- KPIs by Day of Month ---")
print(kpis_by_day.round(2))
print("\n" + "="*80 + "\n")

# --- Understand the trends/seasonality of sales by category, location, month etc ---

# Monthly Sales Trend
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year_month', y='Total_Revenue', data=kpis_by_month, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# Sales by Product Category (Top N)
top_categories = kpis_by_category.sort_values(by='Total_Revenue', ascending=False).head(10)
plt.figure(figsize=(12, 7))
sns.barplot(x='Total_Revenue', y='Product_Category', data=top_categories, palette='magma')
plt.title('Top 10 Product Categories by Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Product Category')
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# Sales by Location
sales_by_location = final_data.groupby('Location')['Sales_amt'].sum().reset_index()
print("--- Sales by Location ---")
print(sales_by_location.round(2))
plt.figure(figsize=(10, 6))
sns.barplot(x='Location', y='Sales_amt', data=sales_by_location, palette='cubehelix')
plt.title('Total Sales Amount by Location')
plt.xlabel('Location')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# --- How number order varies and sales with different days? ---
# Number of Orders and Sales by Day of Week
fig, ax1 = plt.subplots(figsize=(12, 7))

color = 'tab:blue'
ax1.set_xlabel('Day of Week')
ax1.set_ylabel('Number of Orders', color=color)
sns.barplot(x='Day_of_Week', y='Number_of_Orders', data=kpis_by_day_of_week, palette='Blues', ax=ax1)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_title('Number of Orders and Sales by Day of Week')

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Total Revenue', color=color)
sns.lineplot(x='Day_of_Week', y='Total_Revenue', data=kpis_by_day_of_week, color=color, marker='o', sort=False, ax=ax2)
ax2.tick_params(axis='y', labelcolor=color)
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# --- Calculate the Revenue, Marketing spend, percentage of marketing spend out of revenue, Tax, percentage of delivery charges by month. ---

# Monthly Revenue and Marketing Spend
monthly_summary = pd.merge(kpis_by_month[['Year_month', 'Total_Revenue']],
                           marketing_spend.groupby('Year_month')['Total_spend'].sum().reset_index(),
                           how='left', on='Year_month')
monthly_summary.rename(columns={'Total_spend': 'Marketing_Spend'}, inplace=True)

# Calculate percentage of marketing spend out of revenue
monthly_summary['Marketing_Spend_Pct_of_Revenue'] = (monthly_summary['Marketing_Spend'] / monthly_summary['Total_Revenue']) * 100

# Monthly Tax and Delivery Charges
monthly_tax_delivery = final_data.groupby('Year_month').agg(
    Total_Tax=('GST', lambda x: (x * final_data.loc[x.index, 'Quantity'] * final_data.loc[x.index, 'Avg_Price'] * (1 - final_data.loc[x.index, 'Discount_pct'] / 100)).sum()),
    Total_Delivery_Charges=('Delivery_Charges', 'sum')
).reset_index()

monthly_summary = pd.merge(monthly_summary, monthly_tax_delivery, how='left', on='Year_month')

# Calculate percentage of delivery charges out of revenue
monthly_summary['Delivery_Charges_Pct_of_Revenue'] = (monthly_summary['Total_Delivery_Charges'] / monthly_summary['Total_Revenue']) * 100

print("--- Monthly Financial Summary (Revenue, Marketing Spend, Tax, Delivery Charges) ---")
print(monthly_summary.round(2))
print("\n" + "="*80 + "\n")

# --- How marketing spend is impacting on revenue? ---
plt.figure(figsize=(12, 7))
sns.lineplot(x='Year_month', y='Total_Revenue', data=monthly_summary, label='Total Revenue', marker='o')
sns.lineplot(x='Year_month', y='Marketing_Spend', data=monthly_summary, label='Marketing Spend', marker='x')
plt.title('Marketing Spend vs. Total Revenue Over Time')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Marketing_Spend', y='Total_Revenue', data=monthly_summary)
plt.title('Correlation between Marketing Spend and Total Revenue')
plt.xlabel('Marketing Spend')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
print("\nCorrelation between Marketing Spend and Total Revenue: ", monthly_summary['Marketing_Spend'].corr(monthly_summary['Total_Revenue']))
print("\n" + "="*80 + "\n")

# --- Which product was appeared in the transactions? ---
# Top 20 most frequently appearing products (based on unique transactions)
product_appearance = final_data.groupby('Product_Description')['Transaction_ID'].nunique().sort_values(ascending=False).reset_index(name='Number_of_Transactions')

print("--- Top 20 Products by Appearance in Transactions ---")
print(product_appearance.head(20))
plt.figure(figsize=(12, 8))
sns.barplot(x='Number_of_Transactions', y='Product_Description', data=product_appearance.head(20), palette='crest')
plt.title('Top 20 Products by Appearance in Transactions')
plt.xlabel('Number of Transactions')
plt.ylabel('Product Description')
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# --- Which product was purchased mostly based on the quantity? ---
product_quantity = final_data.groupby('Product_Description')['Quantity'].sum().sort_values(ascending=False).reset_index()

print("--- Top 20 Products by Quantity Purchased ---")
print(product_quantity.head(20))
plt.figure(figsize=(12, 8))
sns.barplot(x='Quantity', y='Product_Description', data=product_quantity.head(20), palette='flare')
plt.title('Top 20 Products by Quantity Purchased')
plt.xlabel('Total Quantity Purchased')
plt.ylabel('Product Description')
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")


# --- 3. Performing Customer Segmentation ---

# Calculate RFM values
# Recency: Days since last purchase
# Frequency: Number of unique transactions
# Monetary: Total sales amount
snapshot_date = final_data['Transaction_Date'].max() + dt.timedelta(days=1) # One day after the last transaction

rfm = final_data.groupby('CustomerID').agg(
    Recency=('Transaction_Date', lambda date: (snapshot_date - date.max()).days),
    Frequency=('Transaction_ID', 'nunique'),
    Monetary=('Sales_amt', 'sum')
).reset_index()

print("--- RFM Metrics Sample ---")
print(rfm.head())
print("\n" + "="*80 + "\n")

# --- RFM Histograms ---
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(rfm['Recency'], bins=20, kde=True, color='skyblue')
plt.title('Recency Distribution')
plt.xlabel('Days Since Last Purchase')
plt.ylabel('Number of Customers')

plt.subplot(1, 3, 2)
sns.histplot(rfm['Frequency'], bins=20, kde=True, color='lightcoral')
plt.title('Frequency Distribution')
plt.xlabel('Number of Transactions')
plt.ylabel('Number of Customers')

plt.subplot(1, 3, 3)
sns.histplot(rfm['Monetary'], bins=20, kde=True, color='lightgreen')
plt.title('Monetary Distribution')
plt.xlabel('Total Sales Amount')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# --- RFM Pairplot ---
sns.pairplot(rfm[['Recency', 'Frequency', 'Monetary']])
plt.suptitle('Pairplot of RFM Features', y=1.02)
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# --- Heuristic (Value based, RFM) Segmentation ---
# Define RFM scores (higher score for better performance)
rfm['R_Score'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1], duplicates='drop') # Lower recency = higher score
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 4, labels=[1, 2, 3, 4], duplicates='drop') # Higher frequency = higher score
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 4, labels=[1, 2, 3, 4], duplicates='drop') # Higher monetary = higher score

# Handle potential issues if qcut creates fewer than 4 bins due to duplicate values
# If there are fewer than 4 unique values, qcut might drop labels, leading to errors.
# We'll re-assign labels based on available unique labels if duplicates='drop' is used.
# For simplicity, we'll proceed assuming enough unique values for 4 labels.

rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm['RFM_Sum'] = rfm['R_Score'].astype(int) + rfm['F_Score'].astype(int) + rfm['M_Score'].astype(int)

# Define segments based on RFM_Sum or RFM_Score
def rfm_segment(df):
    if df['RFM_Sum'] >= 10:
        return 'Premium'
    elif df['RFM_Sum'] >= 7:
        return 'Gold'
    elif df['RFM_Sum'] >= 4:
        return 'Silver'
    else:
        return 'Standard'

rfm['Customer_Segment_Heuristic'] = rfm.apply(rfm_segment, axis=1)

print("--- Heuristic Customer Segmentation (RFM) ---")
print(rfm.groupby('Customer_Segment_Heuristic').agg(
    Count=('CustomerID', 'count'),
    Avg_Recency=('Recency', 'mean'),
    Avg_Frequency=('Frequency', 'mean'),
    Avg_Monetary=('Monetary', 'mean')
).round(2))
print("\n" + "="*80 + "\n")

# --- Scientific (Using K-Means) Segmentation ---
# Prepare data for K-Means: Scale RFM features
X = rfm[['Recency', 'Frequency', 'Monetary']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters (Elbow Method)
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Set n_init to suppress warning
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('SSE (Sum of Squared Errors)')
plt.xticks(range(1, 11))
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")

# Based on the elbow plot, choose an optimal K (e.g., 3 or 4)
optimal_k = 3 # This can be adjusted based on visual inspection of the plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
rfm['Cluster'] = kmeans.fit_predict(X_scaled)

# Analyze cluster profiles
cluster_profiles = rfm.groupby('Cluster').agg(
    Count=('CustomerID', 'count'),
    Avg_Recency=('Recency', 'mean'),
    Avg_Frequency=('Frequency', 'mean'),
    Avg_Monetary=('Monetary', 'mean')
).round(2)

print(f"--- K-Means Customer Segmentation (K={optimal_k}) Profiles ---")
print(cluster_profiles)
print("\n" + "="*80 + "\n")

# --- Feature Selection for RFM features ---
# VIF for multicollinearity
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]
print("--- Variance Inflation Factor (VIF) for RFM Features ---")
print(vif_data)
print("\n" + "="*80 + "\n")

# PCA for dimensionality reduction
pca = PCA(n_components=3) # Start with 3 components as we have 3 features
pca.fit(X_scaled)
print("--- PCA Explained Variance Ratio ---")
print(pca.explained_variance_ratio_)
print("Cumulative Explained Variance:", pca.explained_variance_ratio_.cumsum())

# Visualize PCA if 2 components explain enough variance
if pca.explained_variance_ratio_.cumsum()[1] > 0.9: # If first two components explain > 90% variance
    pca_2d = PCA(n_components=2)
    X_pca_2d = pca_2d.fit_transform(X_scaled)
    rfm['PCA1'] = X_pca_2d[:, 0]
    rfm['PCA2'] = X_pca_2d[:, 1]
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=rfm, palette='viridis', s=100, alpha=0.7)
    plt.title('K-Means Clusters in PCA 2D Space')
    plt.xlabel(f'Principal Component 1 ({pca_2d.explained_variance_ratio_[0]*100:.2f}%)')
    plt.ylabel(f'Principal Component 2 ({pca_2d.explained_variance_ratio_[1]*100:.2f}%)')
    plt.legend(title='Cluster')
    plt.tight_layout()
    plt.show()
else:
    print("\n2D PCA visualization not performed as first two components do not explain sufficient variance.")
print("\n" + "="*80 + "\n")


# --- 4. Predicting Customer Lifetime Value (CLV) ---
# Define dependent variable: CLV categories (Low, Medium, High Value)
# Using quantiles of total sales amount for CLV categorization
clv_data = final_data.groupby('CustomerID')['Sales_amt'].sum().reset_index()
clv_data.rename(columns={'Sales_amt': 'Total_Revenue'}, inplace=True)

clv_data['CLV_Category'] = pd.qcut(clv_data['Total_Revenue'], q=3, labels=['Low Value', 'Medium Value', 'High Value'])

print("--- CLV Category Distribution ---")
print(clv_data['CLV_Category'].value_counts())
print("\n" + "="*80 + "\n")

# Merge CLV categories with RFM data for features
clv_features = pd.merge(rfm[['CustomerID', 'Recency', 'Frequency', 'Monetary']], clv_data[['CustomerID', 'CLV_Category']], on='CustomerID')

# Prepare data for Classification Model
X_clv = clv_features[['Recency', 'Frequency', 'Monetary']]
y_clv = clv_features['CLV_Category']

# Scale features
scaler_clv = StandardScaler()
X_clv_scaled = scaler_clv.fit_transform(X_clv)

# F-score (ANOVA) for CLV prediction features
f_scores_clv, p_values_clv = f_classif(X_clv_scaled, y_clv)
f_score_df_clv = pd.DataFrame({'Feature': X_clv.columns, 'F-Score': f_scores_clv, 'P-Value': p_values_clv})
print("--- F-Score for CLV Prediction Features ---")
print(f_score_df_clv.round(4))
print("\n" + "="*80 + "\n")


# Split data
X_train_clv, X_test_clv, y_train_clv, y_test_clv = train_test_split(X_clv_scaled, y_clv, test_size=0.3, random_state=42, stratify=y_clv)

# Train a RandomForestClassifier model
clv_model = RandomForestClassifier(n_estimators=100, random_state=42)
clv_model.fit(X_train_clv, y_train_clv)

# Predict and evaluate
y_pred_clv = clv_model.predict(X_test_clv)

print("--- CLV Prediction Model Evaluation ---")
print("Accuracy:", round(accuracy_score(y_test_clv, y_pred_clv), 4)) # Corrected round() usage
print("\nConfusion Matrix:\n", confusion_matrix(y_test_clv, y_pred_clv))
print("\nClassification Report:\n", classification_report(y_test_clv, y_pred_clv))
print("\n" + "="*80 + "\n")

# --- 5. Cross-Selling (Market Basket Analysis) ---
# Prepare data for Apriori: One-hot encode product descriptions per transaction
basket = (final_data.groupby(['Transaction_ID', 'Product_Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0).set_index('Transaction_ID'))

# Convert quantities to binary (1 if purchased, 0 otherwise)
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)

# Find frequent itemsets
# Adjust min_support based on dataset size and desired granularity
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)

# Generate association rules
# Adjust min_confidence based on desired strength of rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Sort rules by lift for better insights
rules = rules.sort_values(['lift', 'confidence'], ascending=[False, False])

print("--- Top 10 Cross-Selling Association Rules ---")
print(rules.head(10))
print("\n" + "="*80 + "\n")

# --- 6. Predicting Next Purchase Day ---
# Highlight Start: Changes made in this block
# Calculate average days per transaction for repeat customers
# Filter for customers with more than one transaction
repeat_customers = final_data.groupby('CustomerID').filter(lambda x: x['Transaction_ID'].nunique() > 1)

# Sort by customer and transaction date
repeat_customers = repeat_customers.sort_values(by=['CustomerID', 'Transaction_Date'])

# Calculate the difference between consecutive transaction dates for each customer
repeat_customers['Days_Since_Last_Purchase'] = repeat_customers.groupby('CustomerID')['Transaction_Date'].diff().dt.days

# Calculate average days between purchases for each repeat customer
avg_days_between_purchases = repeat_customers.groupby('CustomerID')['Days_Since_Last_Purchase'].mean().reset_index()
avg_days_between_purchases.rename(columns={'Days_Since_Last_Purchase': 'Avg_Days_Between_Purchases'}, inplace=True)

# Define dependent variable with categories (0-30, 30-60, 60-90, 90+ days)
def categorize_next_purchase_day(days):
    if days <= 30:
        return '0-30 Days'
    elif 30 < days <= 60:
        return '30-60 Days'
    elif 60 < days <= 90:
        return '60-90 Days'
    else:
        return '90+ Days'

avg_days_between_purchases['Next_Purchase_Category'] = avg_days_between_purchases['Avg_Days_Between_Purchases'].apply(categorize_next_purchase_day)

print("--- Next Purchase Day Category Distribution ---")
print(avg_days_between_purchases['Next_Purchase_Category'].value_counts())
print("\n" + "="*80 + "\n")

# Prepare data for Classification Model (using RFM features for prediction)
next_purchase_features = pd.merge(rfm[['CustomerID', 'Recency', 'Frequency', 'Monetary']],
                                  avg_days_between_purchases[['CustomerID', 'Next_Purchase_Category']],
                                  on='CustomerID', how='inner') # Use inner join to only include repeat customers

X_next_purchase = next_purchase_features[['Recency', 'Frequency', 'Monetary']]
y_next_purchase = next_purchase_features['Next_Purchase_Category']

# Filter out classes with only one member for stratification
# Get value counts of the target variable
class_counts = y_next_purchase.value_counts()
# Identify classes with only one member
single_member_classes = class_counts[class_counts == 1].index

if not single_member_classes.empty:
    print(f"Warning: The following classes in 'Next_Purchase_Category' have only one member and will be excluded from train_test_split for stratification: {list(single_member_classes)}")
    # Filter out rows corresponding to single-member classes
    mask = ~y_next_purchase.isin(single_member_classes)
    X_next_purchase = X_next_purchase[mask]
    y_next_purchase = y_next_purchase[mask]
    # Check if any data remains after filtering
    if X_next_purchase.empty:
        print("Error: No data remains after filtering out single-member classes. Cannot perform train_test_split.")
        # You might want to exit or skip the model training if no data remains
        # For now, we'll let the subsequent split raise an error if needed, or handle gracefully.
    else:
        print(f"Remaining data points after filtering: {len(X_next_purchase)}")
        print("Updated Next Purchase Day Category Distribution after filtering:")
        print(y_next_purchase.value_counts())

# Ensure there's enough data for splitting after filtering
if len(y_next_purchase.unique()) < 2 or len(y_next_purchase) < 2:
    print("Not enough classes or samples remaining for train_test_split after filtering single-member classes.")
    # Skip model training and evaluation if not enough data
else:
    # Scale features
    scaler_next_purchase = StandardScaler()
    X_next_purchase_scaled = scaler_next_purchase.fit_transform(X_next_purchase)

    # F-score (ANOVA) for Next Purchase Day prediction features
    f_scores_np, p_values_np = f_classif(X_next_purchase_scaled, y_next_purchase)
    f_score_df_np = pd.DataFrame({'Feature': X_next_purchase.columns, 'F-Score': f_scores_np, 'P-Value': p_values_np})
    print("--- F-Score for Next Purchase Day Prediction Features ---")
    print(f_score_df_np.round(4))
    print("\n" + "="*80 + "\n")

    # Split data
    X_train_np, X_test_np, y_train_np, y_test_np = train_test_split(X_next_purchase_scaled, y_next_purchase, test_size=0.3, random_state=42, stratify=y_next_purchase)

    # Train a RandomForestClassifier model
    next_purchase_model = RandomForestClassifier(n_estimators=100, random_state=42)
    next_purchase_model.fit(X_train_np, y_train_np)

    # Predict and evaluate
    y_pred_np = next_purchase_model.predict(X_test_np)

    print("--- Next Purchase Day Prediction Model Evaluation ---")
    print("Accuracy:", round(accuracy_score(y_test_np, y_pred_np), 4)) # Corrected round() usage
    print("\nConfusion Matrix:\n", confusion_matrix(y_test_np, y_pred_np))
    print("\nClassification Report:\n", classification_report(y_test_np, y_pred_np))
# Highlight End: Changes made in this block
print("\n" + "="*80 + "\n")

# --- 7. Perform cohort analysis by defining below cohorts ---
# This part was already covered in section 2: "Understand the retention of customers on month on month basis"
# The 'cust_retention' heatmap and associated print statements provide this analysis.
print("--- Cohort Analysis Summary (from section 2) ---")
print("Customers who started in each month and their retention behavior is shown in the 'Retention by Monthly Cohorts (%)' heatmap.")
print("The month cohort with maximum retention can be identified by looking at the last column (or last few columns) of the 'cust_retention' DataFrame.")
print("Based on the 'Customer Retention by Monthly Cohorts (%)' output, you can visually identify the cohort with the highest retention across periods.")
print("\n" + "="*80 + "\n")

# --- Overall Correlation Matrix for Numerical Features ---
numerical_cols = ['Quantity', 'Avg_Price', 'Discount_pct', 'GST', 'Delivery_Charges', 'Invoice_amt', 'Sales_amt']
plt.figure(figsize=(10, 8))
sns.heatmap(final_data[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key Numerical Features')
plt.tight_layout()
plt.show()
print("\n" + "="*80 + "\n")
```

3. Customer Segmentation
Customer segmentation is fundamental for personalized marketing and improved customer relationships. As every customer is unique and can be targeted in different ways, customer segmentation plays an important role. It helps to understand customer profiles and can be helpful in defining cross-sell, upsell, activation, and acquisition strategies. Two main approaches were used: heuristic (RFM) and scientific (K-Means).

RFM Metrics Calculation
RFM (Recency, Frequency, Monetary) Segmentation is an acronym for a segmentation model based on these three dimensions of customer behavior:

Recency: When was the customer's last order? It means the number of days since a customer made their last purchase. If its a case for a website or an app, this could be interpreted as the last visit day or the last login time. Lower recency indicates a more engaged customer.

Frequency: How often does the customer purchase? This refers to the number of purchases in a given period (e.g., 3 months, 6 months, or 1 year). A higher frequency indicates more engaged customers. Alternatively, this can be defined as the average duration between two transactions.

Monetary: How much money has the customer spent? This is the total amount of money a customer spent in that given period. Big spenders will be differentiated as MVP (Most Valuable Players) or VIP (Very Important Persons). Higher monetary value indicates a more valuable customer.

A sample of the calculated RFM metrics for customers is shown in the code output.

Pairplot: A pairplot visualizes the relationships between each pair of RFM features, including their individual distributions on the diagonal. This helps in understanding correlations and potential clusters.

Heuristic (RFM) Segmentation
Customers are divided into segments based on their RFM scores. Each RFM metric is typically divided into quartiles (or quintiles), assigning a score (e.g., 1 to 4). A combined RFM score (e.g., RFM_Sum) is then used to define segments:

Premium: High RFM scores (e.g., RFM_Sum >= 10). These are your most valuable, recent, frequent, and high-spending customers.

Strategy: Reward loyalty, offer exclusive access to new products, personalized communication, VIP support.

Gold: Mid-to-high RFM scores (e.g., RFM_Sum >= 7). These are valuable customers who might be slightly less recent or frequent than Premium.

Strategy: Encourage higher spending, cross-selling, personalized promotions to move them to Premium.

Silver: Mid-to-low RFM scores (e.g., RFM_Sum >= 4). These customers show some engagement but might need nurturing.

Strategy: Targeted promotions, re-engagement campaigns, personalized recommendations to increase frequency or monetary value.

Standard: Low RFM scores. These might be new customers, one-time buyers, or those at risk of churning.

Strategy: Win-back campaigns, introductory offers, basic re-engagement to reactivate.

The count and average RFM values for each heuristic segment are printed.

Scientific (K-Means) Segmentation
K-Means clustering is an unsupervised machine learning algorithm used to partition n observations into k clusters, where each observation belongs to the cluster with the nearest mean (centroid).

Data Preparation: RFM features are scaled using StandardScaler to ensure that features with larger ranges do not dominate the clustering process.

Elbow Method: This method is used to determine the optimal number of clusters (k). It plots the Sum of Squared Errors (SSE) (also known as inertia) against the number of clusters. The "elbow" point in the plot, where the rate of decrease in SSE sharply changes, suggests a good k.

Clustering: K-Means is applied with the chosen optimal_k.

Cluster Profiles: The average Recency, Frequency, and Monetary values for each cluster are analyzed to understand the characteristics of each scientifically derived segment. This allows for more granular and data-driven segment definitions.

Strategy for each segment: Similar to heuristic segmentation, specific strategies can be defined based on the characteristics of each K-Means cluster (e.g., a cluster with high Recency but low Frequency might be targeted with re-engagement offers).

The cluster profiles are printed in the code output.

Feature Selection for RFM Features
Feature selection helps in understanding the importance of features and can improve model performance by reducing noise and multicollinearity.

Variance Inflation Factor (VIF):

Explanation: VIF measures how much the variance of an estimated regression coefficient is inflated due to multicollinearity (correlation) among the predictor variables. A high VIF (typically > 5 or 10) indicates that a feature is highly correlated with other features, which can lead to unstable model coefficients.

Application: VIF is calculated for the scaled RFM features. If any VIF value is high, it suggests that those features might be redundant or problematic for certain linear models.

Output: The VIF values for Recency, Frequency, and Monetary are printed.

Explanation: PCA is a dimensionality reduction technique that transforms a set of possibly correlated variables into a smaller set of uncorrelated variables called principal components. The first principal component captures the most variance, the second captures the second most, and so on.

Application: PCA is applied to the scaled RFM features.

Explained Variance Ratio: This indicates the proportion of variance in the original data that is explained by each principal component. The cumulative explained variance shows how much total variance is captured by a certain number of components.

Visualization: If the first two principal components explain a substantial amount of the total variance (e.g., > 90%), a 2D scatter plot of the K-Means clusters in the PCA space is generated. This allows for a simplified visualization of the clusters in a lower-dimensional space.

4. Predicting Customer Lifetime Value (CLV)
Customer Lifetime Value (CLV) can be defined as the monetary value of a customer relationship, based on the present value of the projected future cash flows from the customer relationship. CLV is an important concept as it encourages firms to shift their focus from quarterly profits to the long-term health of their customer relationships. It represents an upper limit on spending to acquire new customers and is a crucial element in calculating the payback of advertising spend in marketing mix modeling. In the current world, almost every retailer promotes its subscription, which is further used to understand customer lifetime. Retailers can manage these customers better if they know which customers have a high lifetime value.

Dependent Variable Definition: Total Sales_amt for each customer is calculated. This continuous variable is then transformed into categorical labels: 'Low Value', 'Medium Value', and 'High Value' using quantiles. This means customers are divided into three groups based on their total revenue, with each group representing a CLV category.

Why Predict CLV?
The LTV is an important building block in campaign design and marketing mix management. Although targeting models can help to identify the right customers to be targeted, LTV analysis can help to quantify the expected outcome of targeting in terms of revenues and profits. The LTV is also important because other major metrics and decision thresholds can be derived from it. For example, the LTV is naturally an upper limit on the spending to acquire a customer, and the sum of the LTVs for all of the customers of a brand, known as the customer equity, is a major metric for business valuations. Similarly to many other problems of marketing analytics and algorithmic marketing, LTV modeling can be approached from descriptive, predictive, and prescriptive perspectives.

Model: A RandomForestClassifier is chosen for this prediction task. Random Forest is an ensemble learning method that builds multiple decision trees and merges their predictions to improve accuracy and control overfitting.

Features: The RFM features (Recency, Frequency, Monetary) are used as predictors for CLV.

Feature Selection (F-score):

Explanation: The F-score (ANOVA F-value) is a statistical test used to determine if there are significant differences between the means of two or more groups. In classification, it can be used to assess the discriminative power of numerical features with respect to a categorical target variable. A higher F-score and a lower p-value indicate that the feature is more statistically significant in differentiating between the target classes.

Application: F-scores are calculated for the scaled RFM features against the CLV categories. This helps confirm the relevance of RFM features for CLV prediction.

Output: The F-scores and p-values for each feature are printed.

5. Cross-Selling (Market Basket Analysis)
Market Basket Analysis is a data mining technique used to discover associations between items in large transactional datasets. It helps identify which products customers tend to buy together.

Apriori Algorithm:

Explanation: The Apriori algorithm is a classic algorithm for frequent itemset mining and association rule learning. It identifies frequent individual items in the database and extends them to larger itemsets as long as those itemsets appear sufficiently often in the database.

Process:

One-Hot Encoding: Transaction data is transformed into a one-hot encoded format, where each row represents a transaction and each column represents a product, with a 1 if the product was purchased and 0 otherwise.

Frequent Itemsets: The Apriori algorithm finds itemsets (combinations of products) that meet a minimum support threshold. Support is the proportion of transactions that contain the itemset.

Association Rules: From the frequent itemsets, association rules are generated based on a minimum confidence threshold. An association rule is typically in the form "If {Antecedent} then {Consequent}".

Confidence: The probability that a customer will buy the consequent given that they have bought the antecedent.

Lift: Measures how much more likely the consequent is purchased given the antecedent, relative to its baseline probability. A lift value greater than 1 indicates a positive association (the items are more likely to be bought together than by chance). A lift of 1 means no association, and less than 1 means a negative association.

Application: The code identifies frequent itemsets and then generates association rules, sorted by lift, to highlight the strongest cross-selling opportunities.

Output: The top 10 cross-selling association rules are printed, providing actionable insights for product bundling, recommendation systems, and store layout optimization.
